# Data Cleaning

```python
df['Reviews'].unique()
```

```python
df['Reviews'].str.isnumeric().sum()
```

```python
df[~df['Reviews'].str.isnumeric()]
```

```python
c = df.copy()
```

```python
c.head(2)
```

```python
c = c.drop(c.index[10472])
```

```python
c
```

```python
c[~c['Reviews'].str.isnumeric()]
```

```python
c['Reviews']=c['Reviews'].astype(int)
```

```python
c.info()
```

```python
c['Size'].unique()
```

```python
# 1M = 1000k
```

```python
c['Size'].isnull().sum()
```

```python
c['Size'] = c['Size'].str.replace('M','000')
c['Size'] = c['Size'].str.replace('k','')
c['Size'] = c['Size'].replace('Varies with device',np.nan)
c['Size'] = c['Size'].astype(float)
```

```python
c['Size']
```

```python
c.head(2)
```

```python
c['Installs'].unique()
```

```python
c['Price'].unique()
```

```python
toremove= ['+',',','$']
toclean=['Installs','Price']

for item in toremove:
    for cols in toclean:
        c[cols]=c[cols].str.replace(item,'')
```

```python
c['Price'].unique()
```

```python
c['Installs'].unique()
```

```python
c.head()
```

```python
c['Installs'] = c['Installs'].astype(int)
c['Price'] = c['Price'].astype(float)
```

```python
c.info()
```

```python
c['Last Updated'].unique()
```

```python
c['Last Updated'] = pd.to_datetime(c['Last Updated'])
c['Day'] = c['Last Updated'].dt.day
c['Month'] = c['Last Updated'].dt.month
c['Year'] = c['Last Updated'].dt.year
```

```python
c.head()
```

```python
c.info()
```

```python
c.drop("Unnamed: 0", axis=1,inplace =True)
```

```python
c.head(2)
```

```python
numeric_features = [feature for feature in c.columns if c[feature].dtype != 'O']
categorical_features = [feature for feature in c.columns if c[feature].dtype == 'O']

# print columns
print('We have {} numerical features : {}'.format(len(numeric_features), categorical_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

```python
for col in categorical_features:
    print(c[col].value_counts(normalize=True)*100)
    print('-----------------------------------------')
```

```python

plt.figure(figsize=(15,15))
plt.suptitle('Univariate Analysis of Numerical Features ', fontsize = 20, fontweight = 'bold' , alpha = 0.8 , y = 1.)

for i in range (0, len(numeric_features)):
    plt.subplot(5,3,i+1)
    sns.kdeplot(x=c[numeric_features[i]],shade=True, color ='r')
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
```

```python
plt.figure(figsize=(15,15))
plt.suptitle('Univariate Analysis of Numerical Features ', fontsize = 20, fontweight = 'bold' , alpha = 0.8 , y = 1)
category = ['Type','Content Rating']
for i in range (0,len(category)):
    plt.subplot(2,2,i+1)
    sns.countplot(x=df[category[i]],palette="Set2")
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
```

```python
category = pd.DataFrame(c['Category'].value_counts())    
category.rename(columns = {'Category': 'count'},inplace = True)    
```

```python
category
```

```python
plt.figure(figsize=(15,16))
sns.barplot(x=category.index[:10], y ='count',data = category[:10] , palette = 'hls')
plt.title('Top 10 App Categories')
plt.xticks(rotation=90)
plt.show()
```

```python
df_cat_installs = c.groupby(['Category'])['Installs'].sum().sort_values(ascending = False).reset_index()
df_cat_installs.Installs = df_cat_installs.Installs/1000000000 # converting into billion
df2 = df_cat_installs.head(10)
plt.figure(figsize = (14,10))
sns.set_context("talk")
sns.set_style("darkgrid")

ax = sns.barplot(x = 'Installs' , y = 'Category' , data = df2)
ax.set_xlabel('No. of Installations in Billion')
ax.set_ylabel('')
ax.set_title("Most Popular Categories in Play Store", size = 20)
```

```python
dfa = c.groupby(['Category' , 'App'])['Installs'].sum().reset_index()
dfa = dfa.sort_values('Installs' , ascending = False)
apps = ['GAME' , 'COMMUNICATION', 'PRODUCTIVITY', 'SOCIAL']
sns.set_context("poster")
sns.set_style("darkgrid")

plt.figure(figsize =(40,30))

for i, app in enumerate(apps):
    df2 = dfa[dfa.Category == app]
    df3 = df2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(data = df3 , x='Installs', y ='App')
    plt.xlabel('Installation in Millions')     
    plt.ylabel('')
    plt.title(app,size = 20)

plt.tight_layout()
plt.subplots_adjust(hspace = .3)
plt.show()
```

```python
rating = c.groupby(['Category','Installs','App'])['Rating'].sum().sort_values(ascending = False).reset_index()

toprating_apps = rating[rating.Rating == 5.0]
print("Number od 5 rated apps",toprating_apps.shape[0])
toprating_apps.head(1)
```
